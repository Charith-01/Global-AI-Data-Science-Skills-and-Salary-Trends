```python
def categorize_salary(salary):
    if salary < 80000:
        return "Low"
    elif salary < 120000:
        return "Medium"
    elif salary < 150000:
        return "High"
    else:
        return "Very High"

ai_jobs["salary_level"] = ai_jobs["salary_avg_usd"].apply(categorize_salary)

ai_jobs[["salary_avg_usd", "salary_level"]].head()
```

```python
def create_career_stage(level):
    if level == "Entry":
        return "Entry Level"
    elif level == "Mid":
        return "Mid Level"
    elif level == "Senior":
        return "Senior Level"
    else:
        return "Unknown"

ai_jobs["career_stage"] = ai_jobs["experience_level"].apply(create_career_stage)

ai_jobs[["experience_level", "career_stage"]].head()
```

```python
def remote_friendly(remote_type):
    if remote_type in ["Remote", "Hybrid"]:
        return "Yes"
    elif remote_type == "Onsite":
        return "No"
    else:
        return "Unknown"

ai_jobs["is_remote_friendly"] = ai_jobs["remote_type"].apply(remote_friendly)

ai_jobs[["remote_type", "is_remote_friendly"]].head()
```

```python
def create_role_group(job_title):
    if job_title == "Data Analyst":
        return "Data Analytics"
    elif job_title == "Data Scientist":
        return "Data Science"
    elif job_title == "Machine Learning Engineer":
        return "Machine Learning"
    elif job_title == "MLOps Engineer":
        return "MLOps"
    elif job_title == "AI Researcher":
        return "AI Research"
    elif job_title == "Applied Scientist":
        return "Applied AI / Research"
    else:
        return "Other"

ai_jobs["role_group"] = ai_jobs["job_title"].apply(create_role_group)

ai_jobs[["job_title", "role_group"]].head()
```

```python
jobs_by_year = ai_jobs.groupby("posted_year")["job_id"].count().reset_index()

jobs_by_year.columns = ["posted_year", "job_count"]

jobs_by_year
```

```python
jobs_by_country = ai_jobs.groupby("country")["job_id"].count().reset_index()

jobs_by_country.columns = ["country", "job_count"]

jobs_by_country = jobs_by_country.sort_values(by="job_count", ascending=False)

jobs_by_country
```

```python
jobs_by_role = ai_jobs.groupby("role_group")["job_id"].count().reset_index()

jobs_by_role.columns = ["role_group", "job_count"]

jobs_by_role = jobs_by_role.sort_values(by="job_count", ascending=False)

jobs_by_role
```

```python
salary_by_role = ai_jobs.groupby("role_group")["salary_avg_usd"].mean().reset_index()

salary_by_role.columns = ["role_group", "avg_salary_usd"]

salary_by_role = salary_by_role.sort_values(by="avg_salary_usd", ascending=False)

salary_by_role
```

```python
salary_by_country = ai_jobs.groupby("country")["salary_avg_usd"].mean().reset_index()

salary_by_country.columns = ["country", "avg_salary_usd"]

salary_by_country = salary_by_country.sort_values(by="avg_salary_usd", ascending=False)

salary_by_country
```

```python
salary_by_career_stage = ai_jobs.groupby("career_stage")["salary_avg_usd"].mean().reset_index()

salary_by_career_stage.columns = ["career_stage", "avg_salary_usd"]

salary_by_career_stage = salary_by_career_stage.sort_values(by="avg_salary_usd", ascending=False)

salary_by_career_stage
```

```python
jobs_by_remote_type = ai_jobs.groupby("remote_type")["job_id"].count().reset_index()

jobs_by_remote_type.columns = ["remote_type", "job_count"]

jobs_by_remote_type = jobs_by_remote_type.sort_values(by="job_count", ascending=False)

jobs_by_remote_type
```

```python
salary_by_remote_type = ai_jobs.groupby("remote_type")["salary_avg_usd"].mean().reset_index()

salary_by_remote_type.columns = ["remote_type", "avg_salary_usd"]

salary_by_remote_type = salary_by_remote_type.sort_values(by="avg_salary_usd", ascending=False)

salary_by_remote_type
```

```python
top_skills = skills_demand.groupby("skill")["job_id"].count().reset_index()

top_skills.columns = ["skill", "demand_count"]

top_skills = top_skills.sort_values(by="demand_count", ascending=False)

top_skills
```

```python
skill_category_demand = skills_demand.groupby("skill_category")["job_id"].count().reset_index()

skill_category_demand.columns = ["skill_category", "demand_count"]

skill_category_demand = skill_category_demand.sort_values(by="demand_count", ascending=False)

skill_category_demand
```

```python
skill_level_demand = skills_demand.groupby("skill_level")["job_id"].count().reset_index()

skill_level_demand.columns = ["skill_level", "demand_count"]

skill_level_demand = skill_level_demand.sort_values(by="demand_count", ascending=False)

skill_level_demand
```

```python
analysis_dir = PROCESSED_DATA_DIR / "analysis_tables"
analysis_dir.mkdir(exist_ok=True)

jobs_by_year.to_csv(analysis_dir / "jobs_by_year.csv", index=False)
jobs_by_country.to_csv(analysis_dir / "jobs_by_country.csv", index=False)
jobs_by_role.to_csv(analysis_dir / "jobs_by_role.csv", index=False)
salary_by_role.to_csv(analysis_dir / "salary_by_role.csv", index=False)
salary_by_country.to_csv(analysis_dir / "salary_by_country.csv", index=False)
salary_by_career_stage.to_csv(analysis_dir / "salary_by_career_stage.csv", index=False)
jobs_by_remote_type.to_csv(analysis_dir / "jobs_by_remote_type.csv", index=False)
salary_by_remote_type.to_csv(analysis_dir / "salary_by_remote_type.csv", index=False)
top_skills.to_csv(analysis_dir / "top_skills.csv", index=False)
skill_category_demand.to_csv(analysis_dir / "skill_category_demand.csv", index=False)
skill_level_demand.to_csv(analysis_dir / "skill_level_demand.csv", index=False)

print("Analysis tables saved successfully")
```
