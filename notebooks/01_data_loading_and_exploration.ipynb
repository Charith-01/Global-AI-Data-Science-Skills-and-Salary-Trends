The dataset was checked for missing values, duplicate records, data types, salary statistics, and key categorical values.

- Salary and year columns are stored as numeric data types.
- The dataset covers job market data from 2020 to 2026.
- Remote types include Remote, Hybrid, and Onsite.
- Experience levels include Entry, Mid, and Senior.
- Employment type contains only Full-time roles, so it will not be used for major comparison.
- The main countries are USA, India, Germany, UK, Canada, and Australia.
- The main job titles are MLOps Engineer, AI Researcher, Data Scientist, Applied Scientist, Data Analyst, and Machine Learning Engineer.
- The main industries are Tech, Healthcare, Retail, Finance, and Education.
- The main skill categories are ML, Cloud, and Programming.
- Country trend data shows that AI/Data Science job demand fluctuated from 2020 to 2026, with a recovery in 2026.

```python
def categorize_salary(salary):
    if salary < 80000:
        return "Low"
    elif salary < 120000:
        return "Medium"
    elif salary < 150000:
        return "High"
    else:
        return "Very High"

ai_jobs["salary_level"] = ai_jobs["salary_avg_usd"].apply(categorize_salary)

ai_jobs[["salary_avg_usd", "salary_level"]].head()
```

```python
def create_career_stage(level):
    if level == "Entry":
        return "Entry Level"
    elif level == "Mid":
        return "Mid Level"
    elif level == "Senior":
        return "Senior Level"
    else:
        return "Unknown"

ai_jobs["career_stage"] = ai_jobs["experience_level"].apply(create_career_stage)

ai_jobs[["experience_level", "career_stage"]].head()
```

```python
def remote_friendly(remote_type):
    if remote_type in ["Remote", "Hybrid"]:
        return "Yes"
    elif remote_type == "Onsite":
        return "No"
    else:
        return "Unknown"

ai_jobs["is_remote_friendly"] = ai_jobs["remote_type"].apply(remote_friendly)

ai_jobs[["remote_type", "is_remote_friendly"]].head()
```

```python
def create_role_group(job_title):
    if job_title == "Data Analyst":
        return "Data Analytics"
    elif job_title == "Data Scientist":
        return "Data Science"
    elif job_title == "Machine Learning Engineer":
        return "Machine Learning"
    elif job_title == "MLOps Engineer":
        return "MLOps"
    elif job_title == "AI Researcher":
        return "AI Research"
    elif job_title == "Applied Scientist":
        return "Applied AI / Research"
    else:
        return "Other"

ai_jobs["role_group"] = ai_jobs["job_title"].apply(create_role_group)

ai_jobs[["job_title", "role_group"]].head()
```
