## Data Quality Check Summary

The dataset was checked for missing values, duplicate records, data types, salary statistics, and key categorical values.

Key findings:
- No missing values were found in the main datasets.
- No duplicate rows were found.
- Salary and year columns are stored as numeric data types.
- The dataset covers job market data from 2020 to 2026.
- Remote types include Remote, Hybrid, and Onsite.
- Experience levels include Entry, Mid, and Senior.
- Employment type contains only Full-time roles, so it will not be used for major comparison.
- The main countries are USA, India, Germany, UK, Canada, and Australia.
- The main job titles are MLOps Engineer, AI Researcher, Data Scientist, Applied Scientist, Data Analyst, and Machine Learning Engineer.
- The main industries are Tech, Healthcare, Retail, Finance, and Education.
- The main skill categories are ML, Cloud, and Programming.
- Country trend data shows that AI/Data Science job demand fluctuated from 2020 to 2026, with a recovery in 2026.
